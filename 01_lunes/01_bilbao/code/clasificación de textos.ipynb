# paso 1: tokenizar el texto

```python
f = open(r'notaVirus.txt','r',errors='ignore')
texto = f.read()
```

```python
texto
```

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(texto,'spanish')
tokens=[word.lower() for word in tokens if word.isalpha()] #remover todos los símbolos y pasar a minúscula
```

```python
tokens
```

```python
#frecuencia de palabras
frecuencia = nltk.FreqDist(tokens)
for key, val in frecuencia.items():
    print(str(key) + ':' + str(val))
```

```python
#visualizar tokens
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
frecuencia.plot(30,cumulative=False)
```

# Paso 2: Eliminar las palabras de parada

```python
from nltk.corpus import stopwords
```

```python
tokensLimpios = tokens[:]
for token in tokens:
    if token in stopwords.words('spanish'):
        tokensLimpios.remove(token)
print(tokensLimpios)
```

```python
#frecuencia de palabras
frecuenciaLimpia = nltk.FreqDist(tokensLimpios)
for key, val in frecuenciaLimpia.items():
    print(str(key) + ':' + str(val))
```

```python
frecuenciaLimpia.plot(30,cumulative=False)
```

# paso 3: obtener sinónimos y antónimos

```python
from nltk.corpus import wordnet
sinonimos =[]

for sin in wordnet.synsets('Vaccine'):
    for lema in sin.lemmas():
        sinonimos.append(lema.name())
print(sinonimos)
```

```python
#para español, vamos a reemplazar directamente
sinonimos =['adenavirus','coronavirus']
for ind,sin in enumerate(sinonimos):
    tokenSinonimo = [word.replace(sinonimos[ind],'virus') for word in tokensLimpios]
tokenSinonimo
```

```python
#frecuencia de palabras
frecuenciaLimpiaSinonimos = nltk.FreqDist(tokenSinonimo)
for key, val in frecuenciaLimpiaSinonimos.items():
    print(str(key) + ':' + str(val))
```

```python
frecuenciaLimpiaSinonimos.plot(30,cumulative=False)
```

# derivación regresiva

```python
#derivación regresiva en español
from nltk.stem import SnowballStemmer
stem = SnowballStemmer('spanish')
print(stem.stem('Trabajando'))
print(stem.stem('Trabajo'))
```

```python
tokensRegresivo = [stem.stem(token) for token in tokenSinonimo]
tokensRegresivo
```

```python
frecuenciaregresiva = nltk.FreqDist(tokensRegresivo)
for key, val in frecuenciaregresiva.items():
    print(str(key) + ':' + str(val))
```
